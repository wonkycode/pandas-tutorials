**Single Column Sorting:**

```python
import pandas as pd

data = pd.read_csv("datasets/basketball.csv").dropna(how="all")
```

```python
data.head(3)
```

```python
data.sort_values("birth_year")
```

```python
data.sort_values("birth_year", ascending=False)
```

```python
data.head()
```

```python
data.sort_values("birth_year", ascending=False, inplace=True)
```

```python
data.head()
```

By default, NaN values comes last after sorting. To override that you can use na_position parameter.

**The <code>sort_index()</code> Method:**

```python
data.sort_index(inplace=True)
```

```python
data.head()
```

```python
data.sort_values("birth_year", na_position="first")
```

```python
data.sort_values("birth_year", ascending=False, na_position="first")
```

**Multiple Column Sorting:**

```python
data.sort_index(inplace=True)
```

```python
data.head()
```

```python
data.sort_values(["MIN", "birth_year"])
```

```python
data.sort_values(["MIN", "birth_year"], ascending=False)
```

```python
data.sort_values(["MIN", "birth_year"], ascending=[True, False])
```

```python
data.sort_values(by=["MIN", "birth_year"], ascending=[True, False], inplace=True)
```

```python
data.head()
```
